# Scatter plot

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
cars_data=pd.read_csv("toyota.csv",index_col=0,na_values=["??"])
```

```python
cars_data.dropna(axis=0,inplace=True)
```

```python
cars_data.head()
```

```python
plt.scatter(cars_data['age'],cars_data['price'],c='red')
plt.title("sactter plot")
plt.xlabel('age(m)')
plt.ylabel('price')
plt.show()
```

# Histogram

```python
plt.hist(cars_data['km'],color='green',edgecolor='white')
plt.title('Histogram')
plt.xlabel('km')
plt.ylabel('frequency')
plt.show()
```

# Bar plot

```python
counts=[979,120,12]
fudtype=('petrol','diesel','cng')
index=np.arange(len(fudtype))
```

```python
plt.bar(index,counts,color=['red','blue','cyan'])
plt.title("Bar plot")
plt.xlabel('fudtype')
plt.ylabel('frequency')
plt.xticks(index,fudtype,rotation=90)
plt.show()
```

# Scatter plot

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
cars_data=pd.read_csv("toyota.csv",index_col=0,na_values=["??"])
cars_data.dropna(axis=0,inplace=True)
```

```python
sns.set(style="darkgrid")
sns.regplot(x=cars_data['age'],y=cars_data['price'],fit_reg=False,marker="*")
```

```python
sns.lmplot(x='age',y='price',data=cars_data,fit_reg=False,hue='fudtype',legend=True,palette="Set1")
```

# Histogram

```python
sns.distplot(cars_data['age'])
```

```python
sns.distplot(cars_data['age'],kde=False)
```

# Bar plot

```python
sns.countplot(x='age',data=cars_data)
```

```python
sns.countplot(x='age',data=cars_data,hue='metcolor')
```

# Box and Whiskers

```python
sns.boxplot(y=cars_data['price'])
```

```python
sns.boxplot(x=cars_data['metcolor'],y=cars_data['price'])
```

```python
sns.boxplot(x=cars_data['metcolor'],y=cars_data['price'],data=cars_data,hue='automatic')
```

```python
f,(ax_box,ax_hist)=plt.subplots(2,gridspec_kw={"height_ratios":(.15,.85)})
sns.boxplot(cars_data['price'],ax=ax_box)
sns.distplot(cars_data['price'],ax=ax_hist,kde=False)
```

```python
sns.pairplot(cars_data,kind='scatter',hue='fudtype')
plt.show()
```
